```python
x=[1,2,9,4,5,6,7]
y=[100,80,1,4,60,30,20]
m,c,r,p,std_err=stats.linregress(x,y)
def fun(x):
    return(m*x)+c

plt.scatter(x,y)
p=list(map(fun,x))
plt.scatter(x,y)
plt.plot(x,p)
plt.show()
```

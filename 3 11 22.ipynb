```python
#Group chart
x=np.array([5,7,8,7,2])
y1=[5,7,8,7,2]
y2=[4,11,12,9,6]
y3=[88,10,2,10,87]
plt.bar(x-.2,y1,width=0.4,color="violet")
plt.bar(x+.2,y2,width=0.4,color="blue")

plt.title("first")
plt.xticks(x,[2,4,6,8,10])
plt.legend(["a","b","c"])
plt.xlabel("x axis")
plt.ylabel("y axis")





plt.show()
```

```python
#stack chart
x=["a","b","c","d"]
y1=np.array([5,7,8,7])
y2=np.array([4,11,12,9])
y3=np.array([4,9,10,56])
y4=np.array([12,4,6,10])
plt.bar(x,y1,width=0.4,color="violet")
plt.bar(x,y2,width=0.4,color="yellow",bottom=y1)
plt.bar(x,y3,width=0.4,color="black",bottom=y1+y2)
plt.bar(x,y4,width=0.4,color="blue",bottom=y1+y2+y3)
plt.show()
```

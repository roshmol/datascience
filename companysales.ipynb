```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("company_sales_data.csv")
```

```python
#line style=dotted clr=green show legend at lowerright add a title add a circle marker line width should be 3


x=df["month_number"]
y=df["total_profit"]
plt.plot(x,y,color="green",linestyle="dotted",marker='*',linewidth=3,markerfacecolor="red",ms=10) #ms=marker size
plt.xlabel('month_number')
plt.ylabel('total_profit')
plt.title('company profit graph')
plt.xticks(x,["jan","feb","march","april","May","jun","jraph'uly","aug","sep","oct","nov","dec"])
plt.legend("company profit",loc="lower right")

plt.show()
```

```python
#read all product sales data and show in a multty line plot

x=df["month_number"]
y1=df["facecream"]
y2=df["facewash"]
y3=df["toothpaste"]
y4=df["bathingsoap"]
y5=df["shampoo"]
y6=df["moisturizer"]
df.plot(x="month_number",y=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"])
plt.show()
```

```python
#read tooth paste sales data of each month abd show it using scatter polt with grid lines
x=df["month_number"]

y=df["toothpaste"]

plt.xticks(x,["jan","feb","march","april","May","jun","july","aug","sep","oct","nov","dec"])
plt.scatter(x,y,color="red")
plt.title("toothpaste sales graph")
plt.grid(True)

plt.show()
```

```python
#read face wash and face cream and show it in a bar bargraph
x=df["month_number"]

y1=df["facecream"]
y2=df["facewash"]
plt.xticks(x,["jan","feb","march","april","May","jun","july","aug","sep","oct","nov","dec"])
plt.bar(x-.2,y1,width=0.4,color="violet")
plt.bar(x+.2,y2,width=0.4,color="blue")
plt.xlabel('month_number')
plt.ylabel('profit')
plt.title('company profit graph')
plt.show()
```
